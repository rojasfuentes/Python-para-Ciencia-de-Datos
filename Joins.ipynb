```python
import pandas as pd
consumidores = [("A", "Móvil"), ("B", "Movil"), ("A", "Portatil"), ("A", "Tablet"), ("B", "Tablet"), ("B", "Tablet"), ("C", "Portatil"), ("D","Smartwatch"), ("E", "Consola")]

con_labels = ["Consumidor", "Producto"]

con_df = pd.DataFrame.from_records(consumidores,columns = con_labels)

productores =  [("a", "Móvil"), ("b", "Smartwatch"), ("a", "Tablet"), ("b", "Portatil"), ("c", "Sobremesa"), ("c", "Portatil")]

prod_labels = ["Productor", "Producto"]

prod_df= pd.DataFrame.from_records(productores,columns =prod_labels)

#Creamos dos DT
```

```python
con_df
```

```python
prod_df
```

```python
#merge une las bases de datos 
#How indica el tipo de join que vamos a utilizar
#Para este caso usamos outer
pd.merge(con_df,prod_df,on="Producto", how="outer")
```

```python
#Utilizamos inner en how
#Inner no agrega valores NaN, omite los resultados que no coinciden en las dos tablas 
pd.merge(con_df,prod_df,on="Producto", how="inner")
```

```python
#right indica el orden
#junta los data frames si encuentra la relacion sen el dataframe de la derecha 
pd.merge(con_df,prod_df,on="Producto", how="right")
```

```python
pd.merge(con_df,prod_df,on="Producto", how="left")
```
