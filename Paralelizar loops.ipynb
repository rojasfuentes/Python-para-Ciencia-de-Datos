```python
import pandas as pd
import numpy as np
from joblib import Parallel, delayed
#joblib hace procesos en paralelo 
df = pd.read_csv("C:/Users/Mayo/Documents/base_datos_2008.csv", nrows=10000)
```

```python
#Indica un numero determinado de ciclos a realizar 
#Creamos un subset con nuestras ultimas columnas de nuestra bdd
df_sub = df[['CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay']]
df_sub.head(10)
```

```python
#La función calcula el retaso máximo 
#El if define si hay anys en la fila 
#Luego definimos los nombres de las files
#Devolvemos el nombre de la fila, y la poosición que halla tomado un valor máximo
def retraso_maximo(fila):
    if not np.isnan(fila).any():
        names = ['CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay']
        return names[fila.index(max(fila))]
    else:
        return "None"
```

```python
#Implementamos la función en un for
results = []
for fila in df_sub.values.tolist():
    results.append(retraso_maximo(fila))
```

```python
results
```

```python
#La función parallel depende de dos parametros 1.- El numero de procesadores que queremos 2.- especificamos la función map
#El nombre de la función que queremos aplicar y dond queremos hacerlo 
result = Parallel(n_jobs = 2, backend = 'multiprocessing')(map(delayed(retraso_maximo), df_sub.values.tolist()))
```
