```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/Mayo/Documents/base_datos_2008.csv", nrows=10000)
```

```python
#Creamos uan tabla resumen de los vuelos cancelados 
#Son datos 0 y 1 y cuantas veces aparecen en nuestra bdd
data = np.unique(df.Cancelled, return_counts = True)
data
```

```python
plt.pie(x = data[1], 
        labels = data[0],
        colors = ["Green", "Red"],
       shadow= True,
       startangle= 90,
       radius = 2)
plt.show()

#Para más informacion ver documentación de la función pie
```
