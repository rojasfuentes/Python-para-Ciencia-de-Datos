```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(0)
df = pd.read_csv("C:/Users/Mayo/Documents/base_datos_2008.csv", nrows=100000)

data = np.unique(df.DayOfWeek,return_counts=True) #Tabla resumen dekl no de vuelos de cada semana 
labs= ["Lun", "Mar", "Mie", "Jue", "Vie", "Sab", "Dom"]
data
```

```python
plt.pie(x = data[1],
       #labels= labs,
       radius = .7,
       colors = sns.color_palette("hls",7),
        explode = ( 0,0,0,0.2,0,0,0.1),
        autopct = "%1.1f%%",
       startangle = 90)
plt.legend(loc="upper right",labels = labs)
plt.show
```

```python
plt = sns.barplot( x= labs, y= data[1])
plt.set(xlabel = "Día de la semana", ylabel = "Número de Vuelos")
```
