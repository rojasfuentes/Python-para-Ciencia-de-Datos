```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/Mayo/Documents/base_datos_2008.csv")
```

```python
#Eliminamos datos faltantes
df.dropna(inplace = True, subset =["ArrDelay", "DepDelay", "Distance"])
```

```python
#kde elimina la aproximacion de la densidad
#bins define el numero de cortes del histograma
sns.displot(df["Distance"], kde = False, bins = 100)
```

```python
#Visualizamos dos distribuciones
#xlim indica el limite para visualizarlo mejor
sns.kdeplot(df["ArrDelay"])
sns.kdeplot(df["DepDelay"])
plt.xlim([-300,300])
```

```python
#Nuevo df para ATL, HOU e IND
df2 = df[df["Origin"].isin(["ATL","HOU","IND"])].sample(frac = 1).head(500)
```

```python
#en data especificamos el dataframe que utilizamosw
sns.boxplot(x="DepDelay", y="Origin", data = df2)
plt.xlim([-50,150])

#Como interpretar un boxplot
#La linea que atravieza el cuadrado, indica la mediana 
#La linea antes del primer cuadrado, indica el primer cuartil
#La liena despues del cuadrado es el tercer cuartil 
#Los rombos son outliers que detecta el boxplot
#El rango intercuartilico es el inicio y el final del cuadrado de color 
```
