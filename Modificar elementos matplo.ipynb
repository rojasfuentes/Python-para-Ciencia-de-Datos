```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(0)
df = pd.read_csv("C:/Users/Mayo/Documents/base_datos_2008.csv", nrows=10000)
df= df.sample(frac=1).head(100)
#Las remezclamos y elegimos solo las primeras 100
```

```python
#Creamos el gráfico
#Primer parametro es las columnas de nuestro df 
plt.scatter(x=df.DayofMonth, y=df.ArrDelay, s = df.Distance)
```

```python
#Alpha es un parametro que hace que las burbujas sean traslucidas
plt.scatter(x=df.DayofMonth, y=df.ArrDelay, s = df.Distance, alpha =.3)
```

```python
#CCreamos un filtro para saber que días son fines de semana
#lim es para limite
#Despues son las divisiones en el eje x
#Finalmente es el nombre para determinadas coordenadas 
plt.scatter(x=df.DayofMonth,y=df.ArrDelay, s = df.Distance, alpha =.3, c = df.DayOfWeek.isin([6,7]))
plt.title("Retrasos en EEUU")
plt.xlabel("Dia del Mes") 
plt.ylabel("Retrazo al llegar") 
plt.ylim([0,150])
plt.xticks([0,15,30])
plt.text(x=28,y=120, s="Mi vuelo")
plt.show()
```
